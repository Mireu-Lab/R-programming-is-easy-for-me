```r
install.packages("Stat2Data")
```

```r
library(Stat2Data)
data("Diamonds")
```

```r
ds = Diamonds$PricePerCt

cols = rep('#a8dadc', 9)
cols[3] = '#1d3557'
```

```r
hist(ds, main="캐럿당 가격 분표",
     breaks = 10, las = 1,
     xlab = '캐럿당 가격($)', ylab = '빈도수',
     col = cols, border = '#457b9d')
```
